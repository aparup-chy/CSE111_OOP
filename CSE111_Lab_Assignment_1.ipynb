```python
#Task1

u = input("Sir, please enter your string: ")
cap = 0
low = 0
for i in u:
    if 65<=ord(i)<=90:
        cap+=1
    elif 97<=ord(i)<=122:
        low+=1
if cap>low:
    up = u.upper()
    print(up)
else:
    low = u.lower()
    print(low)
```

```python
#Task2

u = input("Sir, please enter your characters: ")
num = ""
word = ""
mixed = ""
for i in u:
    if 48<=ord(i)<=57:
        num+=i
    elif 65<=ord(i)<=90 or 97<=ord(i)<=122:
        word+=i
    elif 65<=ord(i)<=90 or 97<=ord(i)<=122 or 48<=ord(i)<=57:
        mixed+=i

if u == num:
    print("NUMBER")
elif u == word:
    print("WORD")
else:
    print("Mixed")

```

```python
#Task3

u = input("Sir, please enter your word: ")
out_list = []

for i in range(len(u)):
    if 65<=ord(u[i])<=90:
        out_list.append(i)
out = u[out_list[0]+1:out_list[1]:]

if len(out) != 0:
    print(out)
else:
    print("BLANK")
```

```python
#Task4

user1 = input("Sir, please enter your first word: ")
user2 = input("Sir, please enter your second word: ")
word = ""

for i in user1:
    if i in user2:
        word+=i
for i in user2:
    if i in user1:
        word+=i
        
if len(word)>0:
    print(word)
else:
    print("Nothing in common")
```
